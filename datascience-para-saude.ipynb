## Juntando os dados

```python
def lastest_by_country(data):
    return data.groupby("Country/Region").sum().iloc[:,-4]

```

#### Concatenando dados

```python
informations = [lastest_by_country(confirmed).head(),
lastest_by_country(deaths).head(),
lastest_by_country(recovered).head()]

pd.concat(informations, axis=1)
```

#### Nomeando Colunas

```python
informations = [lastest_by_country(confirmed),
lastest_by_country(deaths),
lastest_by_country(recovered)]

combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

## Letalidade

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['deaths'] + sum_up['recovered'])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

```python
def latest_by_country_at(data, date):
    return data.groupby("Country/Region").sum()[date]


informations = [latest_by_country_at(confirmed, '2/20/20'), latest_by_country_at(deaths, '2/20/20'), latest_by_country_at(recovered, '2/20/20')]
combined_2_20_20 = pd.concat(informations, axis=1)
combined_2_20_20.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_2_20_20.loc["China"]
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
letality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

```python

informations = [latest_by_country_at(confirmed, '2/8/20'), latest_by_country_at(deaths, '2/20/20'), latest_by_country_at(recovered, '2/20/20')]
combined_12 = pd.concat(informations, axis=1)
combined_12.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_12.loc["China"]
letality_rate_3 = sum_up["deaths"] / sum_up["confirmed"]

print(letality_rate_3 * 100)
```

```python
def locate(population, country):
    return population.query(f"Location.str.contains('{country}') and Time==2019", engine='python')


locate(un_population,'Korea')
```

```python
def rename_location(location):
    if location == 'China':
        return 'Mainland China'
    if location == 'Republic of Korea':
        return 'Korea, South'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    if location == 'United States of America':
        return 'US'
    if location == 'China, Hong Kong SAR':
        return 'Hong Kong'
    return location
```

```python

un_population_total['location_for_who'] = un_population_total.index.map(rename_location)
un_population_total.query("location_for_who == 'Korea, South'").head()
```

```python
un_population_total = un_population_total = un_population_total.set_index("location_for_who")
un_population_total.head()
```

```python
combined_expanded = combined.join(un_population_total).sort_values("confirmed", ascending=False).dropna()
combined_expanded.head(20)
```

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 100
combined_expanded.head(10)
```

## Finalizando as comparações

```python
grouped_per_day = pd.DataFrame([confirmed.sum(numeric_only=True)[2:-1], recovered.sum(numeric_only=True)[2:], deaths.sum(numeric_only=True)[2:]], index=["confirmed","recovered","deaths"]).T

grouped_per_day.head(5)
```

#### calculando letalidade dos 5 primeiros dias

```python
grouped_per_day["letality_rate_1"] =  grouped_per_day["deaths"] / grouped_per_day["confirmed"] * 100
grouped_per_day["letality_rate_2"] =  grouped_per_day["deaths"] / (grouped_per_day["deaths"] + grouped_per_day["recovered"]) * 100
grouped_per_day.head()
```

```python
grouped_per_day["letality_rate_1"].plot()
grouped_per_day["letality_rate_2"].plot()
plt.title("Letalidade por casos confirmados no mundo todo")
```

#### Ajustando outlier

```python
grouped_per_day.query('letality_rate_2 < 100.000000',inplace=True)
```

```python
fig = plt.figure(figsize=(10,5))
ax = grouped_per_day['letality_rate_1'].plot(kind='line', color='#fff8bc')
ax.set_facecolor('#1f0a1d')
grouped_per_day['letality_rate_2'].plot(kind='line', color='#ea0063')
plt.legend(['Mortes / Casos confirmados', 'Mortes / (Mortes + Recuperados)'])
plt.title("Letalidade 'pessimista e otimista' no mundo todo, convergindo para o mesmo valor")
plt.grid(color='white', linestyle='-')
plt.ylabel("% letalidade")
plt.xlabel("Fonte: www.alura.com.br e dados Johns Hopkins https://github.com/CSSEGISandData/COVID-19")
plt.show()

```
