## Primeiras analises

#### Lendo arquivo

```python
import pandas as pd

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
last_date = "3/2/20"
confirmed[last_date]

deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

```

## Agrupando dados de acordo com a regiao

```python
confirmed_by_country = confirmed.groupby("Country/Region").sum()
confirmed_by_country.head()
```

#### Plotando grafico de acordo com dados da China

```python
confirmed_by_country.loc["China"][2:].plot()
```

#### Ajustando grafico

```python
import matplotlib.pyplot as plt

#Alterando tamanho
plt.figure(figsize=(10,5));
#Dados
confirmed_by_country.loc["China"][2:].plot();
#Alterando titulo
plt.title("Casos confirmados na China");

```

#### Interpretando graficos

```python
new_cases_china = confirmed_by_country.loc['China'][2:].diff().dropna()
plt.figure(figsize=(10,5));
new_cases_china.plot()
```

```python
plt.figure(figsize=(10,5));

#nao usar grafico de pizza e torta
confirmed_by_country[last_date].plot(kind="pie")
plt.title("Confirmados por pais");

```

#### Ordenando os valores

```python
''' 
10 paises com casos mais confirmados

lembrando que existe um vicio aqui pois paises que nao testaram individuos podem sim ter casos 
ocorrendo mas nao confirmados ainda
'''
plt.figure(figsize=(20,10));
confirmed_by_country[last_date].sort_values(ascending=False)[0:10].plot(kind="bar")
plt.title("Confirmados por pais");



```

## Juntando os dados

```python
def lastest_by_country(data):
    return data.groupby("Country/Region").sum().iloc[:,-4]

```

#### Concatenando dados

```python
informations = [lastest_by_country(confirmed).head(),
lastest_by_country(deaths).head(),
lastest_by_country(recovered).head()]

pd.concat(informations, axis=1)
```

#### Nomeando Colunas

```python
informations = [lastest_by_country(confirmed),
lastest_by_country(deaths),
lastest_by_country(recovered)]

combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

## Letalidade

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']
print(letality_rate_1 * 100)
```
